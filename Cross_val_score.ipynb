```python
from sklearn.datasets import load_iris,load_digits

dataset=load_iris()
dir(dataset)
```

```python
import pandas as pd

df=pd.DataFrame(columns=dataset.feature_names,data=dataset.data)
```

```python
df.head()
```

```python
df["target"]=dataset.target
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC 
from sklearn.model_selection import cross_val_score,train_test_split
```

```python
model1=LogisticRegression()
model2=DecisionTreeClassifier()
model3=RandomForestClassifier(n_estimators=10)
Model4=SVC(kernel="rbf")
```

```python
data=load_digits()
df1=pd.DataFrame(data=data.data,columns=data.feature_names)
```

```python
print(cross_val_score(model1,df.drop("target",axis="columns"),df.target))
print(cross_val_score(model2,df.drop("target",axis="columns"),df.target))
print(cross_val_score(model3,df.drop("target",axis="columns"),df.target))
print(cross_val_score(Model4,df.drop("target",axis="columns"),df.target))
```

```python
print(cross_val_score(model1,df1,data.target))
print(cross_val_score(model2,df1,data.target))
print(cross_val_score(model3,df1,data.target))
print(cross_val_score(Model4,df1,data.target))
```
