```python
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

```python
data=load_iris()
```

```python
cross_val_score(SVC(kernel="linear",C=10,gamma="scale"),data.data,data.target,cv=5)
```

```python
cross_val_score(SVC(kernel="rbf",C=5,gamma="scale"),data.data,data.target,cv=5)
```

```python
cross_val_score(SVC(kernel="poly",C=20,gamma="auto"),data.data,data.target,cv=5)
```

```python
kernal=["poly","rbf","linear"]
C=[5,10,15,20,25,30]
gamma=["scale","auto"]
for i in kernal:
    for j in C:
        for k in gamma:
            out=cross_val_score(SVC(kernel=i,C=j,gamma=k),data.data,data.target,cv=5)
            print(f"kernal:{i},C={j},gamma={k}:{out.mean()}")
```

```python
import pandas as pd
```

```python
from sklearn.model_selection import GridSearchCV
clf=GridSearchCV(SVC(), {
    "kernel":["poly","rbf","linear"],
    "C":[i for i in range(1,11)],
    "gamma":["scale","auto"]
},cv=5,return_train_score=False)

clf.fit(data.data,data.target)
#df=pd.DataFrame(clf.cv_results_)
#df[["param_C","param_gamma","param_kernel","mean_test_score"]]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
model={
    "svm":{
        "model":SVC(),
        "params":{
            "kernel":["poly","rbf","linear"],
            "C":[i for i in range(1,11)],
            "gamma":["scale","auto"]
        }
    },
    "random_forest":{
        "model":RandomForestClassifier(),
        "params":{
            "n_estimators":[i for i in range(1,21)]
        }
    },
    "LogisticRegression":{
        "model":LogisticRegression(),
        "params":{
            "C":[i for i in range(1,11)],
        }
        }
}
```

```python
model.items()
```

```python
score={}
for m,mp in model.items():
    clf1=GridSearchCV(mp["model"],mp["params"],cv=5,return_train_score=False)
    clf1.fit(data.data,data.target)
    score[m]={"best_score":clf1.best_score_,
                "best_params":clf1.best_params_}
```
